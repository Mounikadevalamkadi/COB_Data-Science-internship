```python
tv_shows_df = df[df['type'] == 'TV Show']

# If 'listed_in' contains multiple categories separated by commas, you might need to split them
categories = tv_shows_df['listed_in'].str.split(', ', expand=True).stack()

# Plotting the top 10 categories
plt.figure(figsize=(12, 8))
categories.value_counts().head(10).plot(kind='bar', color='purple')
plt.title('Top 10 TV Show Categories', fontsize=18)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Trend of each type of content Released over the years
last_decade = df[["type", "release_year"]]

# Filter for the last decade (2010 and later)
last_decade = last_decade[last_decade["release_year"] >= 2010]

# Group by release year and count the number of each content type
last_decade_df = last_decade.groupby(["release_year", "type"]).size().reset_index(name="Total Content")

plt.figure(figsize=(10, 6))

# Plotting the count of each content type over the years
count_plot = sns.barplot(x="release_year", y="Total Content", data=last_decade_df, hue="type", palette="pastel")
count_plot.set(title="Trend of each type of content Released over the years")
count_plot.set_xticklabels(count_plot.get_xticklabels(), rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Trend of content on Netflix
last_decade = df[["type", "release_year"]]

# Filter for the last decade (2010 and later)
last_decade = last_decade[last_decade["release_year"] >= 2010]

# Group by release year and count the total content
last_decade_df = last_decade.groupby("release_year")["type"].size().reset_index(name="Total Content")

plt.figure(figsize=(10, 6))

# Plotting the trend of total content over the years
plot_total_content = sns.lineplot(x="release_year", y="Total Content", data=last_decade_df, linewidth=3, color='b')
plot_total_content.set(xlabel="Release Year", ylabel="Total Content", title="Trend of content on Netflix")
plt.show()
```

Result: From this we can state that most content on netflix was released in year 2018, the amount of content is decreasing since then.

```python
# Top 10 Highest content creater Countries on Netflix:
top_10_countries = df['country'].value_counts().head(10)
top_10_countries = pd.DataFrame(top_10_countries).reset_index()
top_10_countries.columns = ['Country', 'Count']

plt.figure(figsize=(10, 8))

# Plotting the bar plot
country_plot = sns.barplot(x='Count', y='Country', data=top_10_countries, palette='pastel')
country_plot.set(title='Top 10 Countries on Netflix Creating Highest Content', xlabel='Count', ylabel='Country')

# Adding labels on the bars
for i, value in enumerate(top_10_countries['Count']):
    country_plot.text(value, i, f'{value}', ha='left', va='center', fontsize=12)

plt.show()
```

Result: Most of the content creators on Netflix is in the United states followed by India and UK.

```python
# Count of Ratings by Movies and Shows
new_catgs = {
    'TV-PG': 'Parental Guidance',
    'TV-MA': 'Mature Audience',
    'TV-Y7-FV': 'Teens',
    'TV-Y7': 'Teens',
    'TV-14': 'Teens',
    'R': 'Mature Audience',
    'TV-Y': 'General Audience',
    'NR': 'Mature Audience',
    'PG-13': 'Teens',
    'TV-G': 'General Audience',
    'PG': 'Teens',
    'G': 'General Audience',
    'UR': 'Mature Audience',
    'NC-17': 'Mature Audience'
}

df['rating'] = df['rating'].replace(new_catgs)

plt.figure(figsize=(10, 6))

# Plotting the count of Rating by Movies and Shows
sns.countplot(x="rating", data=df, palette="pastel", order=df['rating'].value_counts().index)

plt.title("Count of Ratings by Movies and Shows")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

Result: From this we can state that most the netflix content is for Mature audience.

```python
# Number of Movies and TV Shows Released by Year (Top 15 Years)
df_15 = df.groupby(['type', 'release_year']).size().unstack().transpose().fillna(0).sort_values(by='release_year', ascending=False)

# Plotting the top 25 years with a stacked bar plot
df_15.head(15).plot.bar(stacked=True, figsize=(17, 12))

plt.title('Number of Movies and TV Shows Released by Year (Top 15 Years)')
plt.xlabel('Year')
plt.ylabel('Number of Content')
plt.show()
```

Result:From this we can state that Highest number of movies were released in 2018 and then after there is a downfall trend with movie releases by increasing year.

```python
# Most Frequent Content Type created by Directors:
data = df[df['director'] != 'Unknown']
director_type_counts = data.groupby(['director', 'type']).size().unstack().fillna(0)

# Find the most frequent type for each director
director_most_frequent_type = director_type_counts.idxmax(axis=1)

plt.figure(figsize=(10, 6))

# Plotting the count of the most frequent type for directors
sns.countplot(x=director_most_frequent_type, palette='pastel')
plt.title('Most Frequent Content Type Created by Directors')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 Most Frequent Listed_in Categories for Directors in Movies
data = df[df['director'] != 'Unknown']
movies_data = data[data['type'] == 'Movie']

# Find the most frequent listed_in category for each director in movies
director_movie_category_counts = movies_data.groupby(['director', 'listed_in']).size().unstack().fillna(0)
director_most_frequent_category = director_movie_category_counts.idxmax(axis=1)

# Select the top 5 most frequent categories
top_10_categories = director_most_frequent_category.value_counts().head(10)

plt.figure(figsize=(12, 8))

# Plotting the count of the top 5 most frequent listed_in categories for directors in movies
sns.barplot(x=top_10_categories.index, y=top_10_categories.values, palette='pastel')
plt.title('Top 10 Most Frequent Listed_in Categories for Directors in Movies')
plt.xlabel('Listed_in Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Top 15 Listed_in Categories for TV Shows with 3 or More Seasons

# Convert 'duration' column to numeric (extracting the numeric part)
df['duration_numeric'] = pd.to_numeric(df['duration'].str.extract('(\d+)', expand=False), errors='coerce')

# Selecting TV Shows with 3 or more seasons
shows_with_3_or_more_seasons = df[(df['type'] == 'TV Show') & (df['duration_numeric'] >= 3)]['listed_in'].sort_values()

# Selecting the top 10 listed_in categories
top_15_categories = shows_with_3_or_more_seasons.value_counts().head(15)

plt.figure(figsize=(12, 8))

# Plotting the distribution of the top 10 listed_in categories for TV Shows with 3 or more seasons
sns.barplot(y=top_10_categories.index, x=top_10_categories.values, palette='pastel')
plt.title('Top 15 Listed_in Categories for TV Shows with 3 or More Seasons')
plt.xlabel('Count')
plt.ylabel('Listed_in Category')
plt.show()
```

```python
# Monthly Release Patterns Over the Years
df['date_added'] = pd.to_datetime(df['date_added'])  # Convert 'date_added' to datetime
df['month'] = df['date_added'].dt.month_name()
df['year'] = df['date_added'].dt.year

# Group by year and month and create a heatmap
df_h = df.groupby(['year', 'month']).size().unstack().fillna(0)
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(15, 10))
sns.heatmap(df_h[month_order], cmap='viridis', annot=True, fmt='g', cbar_kws={'label': 'Number of Releases'})
plt.title('Monthly Release Patterns Over the Years')
plt.show()
```
